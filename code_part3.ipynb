```python
'''This notebook implements the object tracking algorithm based on Harris corner detection, 
feature cleaner, and iterative Lucas Kanade optic flow (ILK) in pyramid fashion.'''

import numpy as np
import cv2
from sklearn.cluster import KMeans
import argparse
import time
```

```python
def downsample(img):
    size = (int(np.round(img.shape[1]/2)), int(np.round(img.shape[0]/2)))
    sigma = 1.0
    g = cv2.getGaussianKernel(9, sigma)
    G = g * np.transpose(g)
    filtered_img = cv2.filter2D(img, -1, G)
    img_down = cv2.resize(filtered_img, size)
    return img_down

def upsample(img):
    size = (img.shape[1]*2, img.shape[0]*2)
    sigma = 1.0
    g = cv2.getGaussianKernel(9, sigma)
    G = g * np.transpose(g)
    img_up = cv2.resize(img, size)
    img_filtered = cv2.filter2D(img_up, -1, G)
    return img_up
```

```python
def cal_derivative(img1, img2):
    
    dx1 = cv2.Sobel(img1,cv2.CV_64F,1,0,ksize=3)
    dy1 = cv2.Sobel(img1,cv2.CV_64F,0,1,ksize=3)
    dx2 = cv2.Sobel(img2,cv2.CV_64F,1,0,ksize=3)
    dy2 = cv2.Sobel(img2,cv2.CV_64F,0,1,ksize=3)
    ix = (dx1 + dx2) / 2
    iy = (dy1 + dy2) / 2
    
    sigma = 1.0
    k = cv2.getGaussianKernel(9, sigma)
    g = k * np.transpose(k)
    imgFiltered1 = cv2.filter2D(img1, -1, g)
    imgFiltered2 = cv2.filter2D(img2, -1, g)
    it = imgFiltered1 - imgFiltered2
    
    return ix, iy, it
```

```python
def clustering(xs, ys):
    selected_x = []
    selected_y = []
    num_clusters = 10
    pred = KMeans(n_clusters=num_clusters, random_state=9).fit_predict(np.transpose([xs, ys]))
    for group in range(int(np.max(pred)-np.min(pred))):
        group_x = []
        group_y = []
        ds = []
        for i in range(len(pred)):
            if pred[i]==group:
                group_x.append(xs[i])
                group_y.append(ys[i])
        group_x = np.array(group_x)
        group_y = np.array(group_y)
        center_x = np.average(group_x)
        center_y = np.average(group_y)
        
        for i in range(len(group_x)):
            d = (group_x[i]-center_x)**2+(group_y[i]-center_y)**2
            ds.append(d)
        
        selected_x.append(group_x[np.argmin(ds)])
        selected_y.append(group_y[np.argmin(ds)])
    
    return (np.array(selected_x), np.array(selected_y))    
```

```python
def harris_corner(img):
    ix = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=3)
    iy = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=3)
    
    k = 0.05
    sigma = 1.0
    g = cv2.getGaussianKernel(3, sigma)
    G = g * np.transpose(g)
    sum_filter = np.ones((3, 3))
    ixx = cv2.filter2D(ix**2, -1, G)
    iyy = cv2.filter2D(iy**2, -1, G)
    ixy = cv2.filter2D(ix*iy, -1, G)
    sxx = cv2.filter2D(ixx, -1, sum_filter)
    syy = cv2.filter2D(iyy, -1, sum_filter)
    sxy = cv2.filter2D(ixy, -1, sum_filter)
    det = sxx*syy-sxy**2
    trace = sxx + syy
    response = det - k*(trace**2)
    max_response = np.max(response)
    features = np.where(response>=1e-2*max_response)
    filtered_x = []
    filtered_y = []
    for i in range(len(features[0])):
        if i==0:
            filtered_x.append(features[0][i])
            filtered_y.append(features[1][i])
        else:
            bound = 0 if len(filtered_x)<=20 else (len(filtered_x)-20)
            flag = True
            for j in range(bound, len(filtered_x)):
                d = (features[0][i]-filtered_x[j])**2 + (features[1][i]-filtered_y[j])**2
                if d<=25:
                    flag = False
            if flag:
                filtered_x.append(features[0][i])
                filtered_y.append(features[1][i])
    the_feature = clustering(filtered_x, filtered_y)
    return the_feature
```

```python
def LK_optic_flow(img1, img2, features=None):
    U = []
    V = []
    U_map = np.zeros(img1.shape)
    V_map = np.zeros(img1.shape)
    window_size = (5, 5)
    offset_x = int(((window_size[1]-1)/2))
    offset_y = int(((window_size[0]-1)/2))
    sum_kernel = np.ones(window_size)
    ix, iy, it = cal_derivative(img1, img2)
    
    if features is None:
        features = harris_corner(img1)
        
    for i in range(len(features[0])):
        center_x, center_y = features[0][i], features[1][i]
        if center_x < offset_x:
            center_x = offset_x
        if center_x > img1.shape[1] - (offset_x+1):
            center_x = img1.shape[1] - offset_x
        if center_y < offset_y:
            center_y = offset_y
        if center_y >img1.shape[0] - (offset_y+1):
            center_y = img1.shape[0] - offset_y
        pix = ix[int(center_x-offset_x):int(center_x+offset_x), int(center_y-offset_y):int(center_y+offset_y)]
        piy = iy[int(center_x-offset_x):int(center_x+offset_x), int(center_y-offset_y):int(center_y+offset_y)]
        pit = it[int(center_x-offset_x):int(center_x+offset_x), int(center_y-offset_y):int(center_y+offset_y)]
        X = np.reshape(pix, (1, -1)).squeeze()
        Y = np.reshape(piy, (1, -1)).squeeze()
        b = np.reshape(pit, (1, -1)).squeeze()
        AT = np.array([X, Y])
        A = np.transpose(AT)
        ATA = np.matmul(AT, A)
        ATb = np.matmul(AT, b)
        ATA_inv = np.linalg.pinv(ATA)
        x = np.matmul(ATA_inv, ATb)
        U.append(x[0])
        V.append(x[1])
        
    for i in range(len(features[0])):
        xs, ys = np.clip(features[0], 0, img1.shape[0]-1), np.clip(features[1], 0, img1.shape[1]-1)
        U_map[xs[i], ys[i]] = U[i]
        V_map[xs[i], ys[i]] = V[i]
        
    return features, U_map, V_map
```

```python
def LK_iterate(features, img1, img2, u_old, v_old):
    u_old = np.round(u_old)
    v_old = np.round(v_old)
    u_new = np.zeros(u_old.shape)
    v_new = np.zeros(v_old.shape)
    window_size = (5, 5)
    offset_y = int((window_size[0]-1)/2)
    offset_x = int((window_size[1]-1)/2)
    for count in range(len(features[0])):
        i, j = features[0][count], features[1][count]
        if j < offset_x:
            j = offset_x
        if j > img1.shape[1] - (offset_x+1):
            j = img1.shape[1] - offset_x - 1
        if i < offset_y:
            i = offset_y
        if i >img1.shape[0] - (offset_y+1):
            i = img1.shape[0] - offset_y - 1
        patch1 = img1[i-offset_y:i+offset_y+1, j-offset_x:j+offset_x+1]
        left = (j-offset_x)+u_old[i, j]
        right = (j+offset_x)+u_old[i, j]
        up = (i-offset_y)+v_old[i, j]
        down = (i+offset_y)+v_old[i, j]
        if left < 0:
            left = 0
            right = window_size[1]-1
        if right > (img1.shape[1]-1):
            left = img1.shape[1]-5
            right = img1.shape[1]-1
        if up < 0:
            up = 0
            down = window_size[0]-1
        if down > (img1.shape[0]-1):
            up = img1.shape[0]-5
            down = img1.shape[0]-1        
        patch2 = img2[int(up):int(down+1), int(left):int(right+1)]
        pix, piy, pit = cal_derivative(patch1, patch2)
        X = np.reshape(pix, (1, -1)).squeeze()
        Y = np.reshape(piy, (1, -1)).squeeze()
        b = np.reshape(pit, (1, -1)).squeeze()
        AT = np.array([X, Y])
        A = np.transpose(AT)
        ATA = np.matmul(AT, A)
        ATb = np.matmul(AT, b)
        ATA_inv = np.linalg.pinv(ATA)
        x = np.matmul(ATA_inv, ATb)
        u_new[i, j], v_new[i, j] = x[0], x[1]
    
    return (u_old+u_new), (v_old+v_new)
```

```python
def LK_pyramid(f, img1, img2, num_layers, num_iter):
    pyramid1 = []
    pyramid2 = []
    current_layer1 = img1
    current_layer2 = img2
    pyramid1.append(img1)
    pyramid2.append(img2)
    
    for i in range(num_layers-1):
        current_layer1 = downsample(current_layer1)
        current_layer2 = downsample(current_layer2)
        pyramid1.append(current_layer1)
        pyramid2.append(current_layer2)
    
    pyramid1 = pyramid1[::-1]
    pyramid2 = pyramid2[::-1]
    
    for i in range(num_iter-1):
        f = ((f[0]/2).astype(int), (f[1]/2).astype(int))
    
    f, u, v = LK_optic_flow(pyramid1[0], pyramid2[0], f)
    for i in range(num_iter):
        u, v = LK_iterate(f, pyramid1[0], pyramid2[0], u, v)
    
    for i in range(1, num_layers):
        u = upsample(u)*2
        v = upsample(v)*2
        f = (f[0]*2, f[1]*2)
        for i in range(num_iter):
            u, v = LK_iterate(f, pyramid1[i], pyramid2[i], u, v)
        
    return f, u, v
```

```python
def estimate(features, u, v):
    # normalization
#     u[np.where(u==0)] = 1e-5
#     v[np.where(v==0)] = 1e-5
#     threshold = 1e-3
#     for i in range(u.shape[0]):
#         for j in range(u.shape[1]):
#             if u[i, j] > threshold:
#                 u[i, j] = 1
#             elif u[i, j] < -threshold:
#                 u[i, j] = -1
#             else:
#                 u[i, j] = 0
#             if v[i, j] > threshold:
#                 v[i, j] = 1
#             elif v[i, j] < -threshold:
#                 v[i, j] = -1
#             else:
#                 v[i, j] = 0
    u = np.clip(u, 0, 4).astype(int)
    v = np.clip(v, 0, 4).astype(int)
    new_xs = []
    new_ys = []
    for i in range(len(features[0])):
#         print(i, features[0][i], features[1][i], u[features[0][i]][features[1][i]], v[features[0][i]][features[1][i]])
        new_x = features[0][i] + v[features[0][i]][features[1][i]]
        new_y = features[1][i] + u[features[0][i]][features[1][i]]
        if new_x >= u.shape[0]:
            new_x = u.shape[0]-1
        if new_y >= u.shape[1]:
            new_y = u.shape[1]-1
        new_xs.append(int(new_x))
        new_ys.append(int(new_y))
        
    return (np.array(new_xs), np.array(new_ys))
```
